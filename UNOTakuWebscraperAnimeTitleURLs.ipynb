```python
# -*- coding: utf-8 -*-
import csv, numpy as np, pandas as pd, requests, gc, re, urllib3, json
from bs4 import BeautifulSoup as bs

try:
    animeTitleURLsFileName = "UNOTaku_AnimeTitles_URLs_Scraped_Part2.csv"    
    csvAnimeTitleURLs = open(animeTitleURLsFileName, mode ='w') # mode = ['w','a','w+']
    counter = 0
    myAnimeListURLs = 'drive/MyDrive/webscraper_URLs_Part2_UNOTaku_05_2021.txt'
    csvMyAnimeListURLs = open(myAnimeListURLs, mode='r')

    lines = csvMyAnimeListURLs.readlines()
    lenLines = len(lines)
    print(f"lenLines: {lenLines}\n")
    
    csvAnimeTitleURLs.write("URLs of Anime Titles:\n")
    # looking thru soup for all titles derived by 'tbody > tr > td > a'.text 
    # for each element. Then iterate and scrape all titles' URL's.
    for j in range(0,len(lines),1): 
        line = lines[j]
        print(f"I am passing to double for loop... I'm Walkin' here!!\n")
        url = line.replace("\n", "")
        print(f"Element URL: {url}\n")
        soup = bs(requests.get(url).text, 'html.parser')
        # print(f"soup: {soup}\n") 
        arefs = soup.find_all('a', class_='hoverinfo_trigger fw-b fl-l', href=True)
        for a in arefs: 
            href = a['href']
            print(f"href is {href}\n")
            csvAnimeTitleURLs.write("{}\n".format(href))
            print(f"href element no. {counter} has been written to file.\n")
            counter= counter + 1        
        
    csvAnimeTitleURLs.close()
    csvMyAnimeListURLs.close()
    gc.collect()
    
except IOError: 
       print(f"Failure to properly close/read file (IO). Please check stack trace and try again.\n")
except EOFError:
       print(f"Failure to properly read/parse file properly or has reached EOF.Please check stack trace and try again.\n")
except IndentationError: 
       print(f"Failure to properly indent. Please check stack trace and try again.\n")
except ImportError: 
       print(f"Import has failed. Please check imports and try again.\n")
except Exception: 
       print(f"General Exception has occurred. Please check stack trace and try again.\n")
except Error: 
       print(f"General Error has occurred. Please check stack trace and try again.\n")

   





 


```
