```python
import re, urllib3, csv, numpy as np, pandas as pd, requests, gc
from bs4 import BeautifulSoup as bs

try:    
    listURLTitles = []
    for i in range(1,101, 1):
        listURLTitles.append(f"https://www.imdb.com/list/ls057823854/?sort=list_order,asc&st_dt=&mode=detail&page={i}")
    print(f"listURLTitles: {listURLTitles}\n")
    fileName = "webScraper_MovieTitles_List_05_11_2021.csv"

    # Console Input-Based Example of Web Scraping/ File Writing of this form:
    # nameRawFile = raw_input("Insert the name of file to append/create: ")
    # while type(nameRawFile) != String: 
    #       nameRawFile = raw_input("Insert the name of file to append/create: ") 
    # fileTypeExtension = raw_input("Insert the file extension (.fileType) that you wish to use: ")
    # while type(fileTypeExtension ) != String: 
    #       fileTypeExtension  = raw_input("Insert the file extension (.fileType) that you wish to use: ")
    # fileName = ''.join(nameRawFile, fileTypeExtension)

    csvfile = open(fileName, mode ='w') # mode = ['w','a','w+']
    # looking thru soup for all titles derived by 'ul > li > i > a'.text 
    # for each element. Then iterate and scrape all titles.
    counter = 0
    for j in range (0,len(listURLTitles), 1): 
        url = listURLTitles[j]
        print(f"Element URL: {url}\n")
        soup = bs(requests.get(url).text, 'html.parser')
        # print(f"soup: {soup}\n") 
        # 'a', class_='hoverinfo_trigger fw-b fl-l': myAnimeListScraper; NOT MOVIE TITLE
        for el in soup.find_all('h3', class_='lister-item-header'): 
            print(f"Title Name is {el.text}\n")
            print(f"Successfully written element no. ({j}, {counter%100}) to file...\n")
            aref = el.find('a')
            csvfile.write("{}\n".format(aref.text))
            counter += 1
    csvfile.close()
except IOError: 
       print(f"Failure to properly close/read file (IO). Please check stack trace and try again.\n")
except EOFError:
       print(f"Failure to properly read/parse file properly or has reached EOF.Please check stack trace and try again.\n")
except IndentationError: 
       print(f"Failure to properly indent. Please check stack trace and try again.\n")
except ImportError: 
       print(f"Import has failed. Please check imports and try again.\n")
except Exception: 
       print(f"General Exception has occurred. Please check stack trace and try again.\n")
except Error: 
       print(f"General Error has occurred. Please check stack trace and try again.\n")

   
```
