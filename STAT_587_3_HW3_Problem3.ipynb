```
import pandas as pd, numpy as np, scipy.stats, scipy as sci, math,statistics
from google.colab import drive, files, auth

drive.mount('/content/drive', force_remount = True)
ABS_PATHNAME='drive/MyDrive/Colab Notebooks/STAT587-03/Module1/data.csv' 
#all *.csv files in Module1 Folder expected to be of this form (Normal, Binom, Pois, etc.)
#Estimators and confidence interval algorithms in Module 2...
df = pd.DataFrame(pd.read_csv(ABS_PATHNAME)) #Read in the *.csv file
n = int (df.shape[0]) #(a.) Sample Mean
print(f"n:{n}")
Im_X = df["y"] #Data Set
print(f"Image DF\nindex:    value\n {Im_X}") 
mean = statistics.mean(Im_X) #(b.) Sample Mean
median = statistics.median(Im_X) # Sample Median
mode = statistics.mode(Im_X) # Sample Mode
pstdev = statistics.pstdev(Im_X) #Pop. STD_DEV
pvar = statistics.pvariance(Im_X) #Pop. Variance

sample_stdev = statistics.stdev(Im_X) # Sample STD_DEV
sample_var = statistics.variance(Im_X) # Sample Variance
print(f"mean:{mean}\nmedian:{median}\nmode:{mode}\npop_stdev:{pstdev}\npop_var:{pvar}\nsample_stdev:{sample_stdev}\nsample_var:{sample_var}")
quantileList = [25, 50, 75, 97.5, 2.5]
percentileArr =  np.percentile(df.y, quantileList)
print(f"quantileList: {quantileList}")
print(f"percentileArr: {percentileArr}\n")




```
